## 1.1 Scatter and Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas #Library for data manipulation.

data = pandas.read_csv("expDat.txt", sep=",", index_col=0)

plt.figure()
data.plot(figsize=(10,10))
plt.title("Line plot")
plt.xlabel("Time index")
plt.ylabel("Observation index")
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix

plt.figure()
scatter_matrix(data.ix[:, 0:5], alpha=0.5, figsize=(10,10))
plt.show()
```

```python
#Counting the mean vector with a built-in function
mean_vector = data.mean()

#Centering the data.
centered_data = data.values - mean_vector.values

#Calculating the length of the matrix.
length = centered_data.shape[0]
#print(length)

#Calculating the covariance matrix using matrix multiplication.
covariance_matrix = np.dot(centered_data.T, centered_data) / length
#print(covariance_matrix.shape)

plt.figure()

plt.subplot(121)
plt.imshow(covariance_matrix, "hot")
plt.title("Heatmap with counting covariance")
plt.colorbar()

plt.subplot(122)
plt.imshow(data.cov(), "hot")
plt.title("Heatmap with built-in cov function")
plt.colorbar()
plt.show()
```
