## 1.1 Scatter and Line Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas

data = pandas.read_csv("expDat.txt", sep=",", index_col=0)

data.plot(figsize=(10,10))
plt.title("Line plot")
plt.xlabel("Time index")
plt.ylabel("Observation index")
plt.show()
```

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(data.ix[:, 0:5], alpha=0.5,figsize=(10,10))
plt.show()
```

## 1.2 Plotting 3d

```python
#textarray = np.genfromtxt("pca-data-3d.txt", delimiter=",", skip_header=1)
#np.reshape(textarray, textarray.shape + (1,))


from pandas.tools.plotting import scatter_matrix
data = pandas.read_csv("pca-data-3d.txt", sep=",", index_col=0)

scatter_matrix(data, alpha=0.2, figsize=(6,6), diagonal='kde')

plt.show()
```

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D


fig = pyplot.figure()
ax = Axes3D(fig)
txt = np.genfromtxt("pca-data-3d.txt", delimiter=",", skip_header=1)
np.reshape(txt, txt.shape + (1,))
ax.scatter(txt[:, [0]], txt[:, [1]], txt[:, [2]], s=50)
plt.show()
```

We can observe that the points describe a plane wave.
